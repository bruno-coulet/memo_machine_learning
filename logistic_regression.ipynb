```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 100)
# fonction sigmoide
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

y = sigmoid(x)

plt.plot(x, y, label='Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.title('Fonction Sigmoid')

x_classe_0 = np.linspace(-10,-5,5)
x_classe_1 = np.linspace(5,10,5)
y_classe_0 = np.zeros(5)
y_classe_1 = np.ones(5)

plt.scatter(x_classe_0, y_classe_0, color='red', label='Classe 0')
plt.scatter(x_classe_1, y_classe_1, color='blue', label='Classe 1')
plt.legend()
```

```python
# fonction linéaire
def linear(x):
    return 2*x+3

y = linear(x)

plt.plot(x, y, label='Fonction linéaire')
plt.xlabel('x')
plt.ylabel('linaire(x)')
plt.title('Fonction Linéaire')
# plt.legend()

x_classe_0 = np.linspace(-10,-5,5)
x_classe_1 = np.linspace(5,10,5)
y_classe_0 = np.infty(5)
y_classe_1 = np.infty(5)

plt.scatter(x_classe_0, y_classe_0, color='red', label='Classe 0')
plt.scatter(x_classe_1, y_classe_1, color='blue', label='Classe 1')
```
