```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 100) # Generate 100 points from -10 to 10
```

|fonction sigmoïde |
|-|
|transforme une droite en sigmoïde|
|restreint l'axe y à l'intervalle [0, 1]|
|prend une valeur réelle et retourne une probabilité|
|$$\text{p}=\frac{e^\text{log(odds)}}{1 + e^\text{log(odds)}}=\frac{1}{1+e^{-z}}$$|
|logarithme naturel = logarithme népérien<br>c’est-à-dire en base e (avec $e≈2.718$)|

```python
# fonction sigmoide
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

y = sigmoid(x)

plt.plot(x, y, label='Fonction Sigmoïde')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.title('Fonction Sigmoïde')

x_classe_0 = np.linspace(-10,-5,5)
y_classe_0 = np.zeros(5)
plt.scatter(x_classe_0, y_classe_0, color='red', label='Classe 0')
x_classe_1 = np.linspace(5,10,5)
y_classe_1 = np.ones(5)
plt.scatter(x_classe_1, y_classe_1, color='blue', label='Classe 1')

plt.legend()
```

```python
# fonction linéaire
def linear(x):
    return 2*x+3

y = linear(x)

plt.plot(x, y, label='Fonction linéaire')
plt.xlabel('x')
plt.ylabel('linéaire(x)')
plt.title('Fonction Linéaire')
# plt.legend()



# # 5 Points de classes 0 et 1, avec un tableau de 5 valeurs infinies
# x_classe_0 = np.linspace(-10, -5, 5)
# y_classe_0 = np.full(5, np.inf)

# x_classe_1 = np.linspace(5, 10, 5)
# y_classe_1 = np.full(5, np.inf)

# plt.scatter(x_classe_0, y_classe_0, color='red', label='Classe 0')
# plt.scatter(x_classe_1, y_classe_1, color='blue', label='Classe 1')
```
